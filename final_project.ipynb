# Телеком — задача проекта

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

<font size="5">**Описание услуг**

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line, «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).  
Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).
За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.
    
<font size="5">**Описание данных**

Данные состоят из файлов, полученных из разных источников:
    
- `contract_new.csv` — информация о договоре;
- `personal_new.csv` — персональные данные клиента;
- `internet_new.csv` — информация об интернет-услугах;
- `phone_new.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

```python
# snake register
def camel_to_snake(word, sep='_'):
    snake_register = ''
    i = 0
    while i <= len(word)-1:
        if word[i].isupper() and i != len(word)-1: 
            if word[i+1].isupper():
                snake_register += sep + word[i].lower() + word[i+1].lower()
                i += 2
            else:
                snake_register += sep + word[i].lower()
                i += 1
        elif word[i].isupper(): 
            snake_register += sep + word[i].lower()
            i += 1
        else:
            snake_register += word[i]
            i += 1
    if snake_register[0] == sep:
        return snake_register[1:]
    else:
        return snake_register

for table in data: 
    data[table].columns = list(map(camel_to_snake, data[table].columns))
```

## Создание модели

```python
# еще раз посмотрим на данные (по возможности булевы признаки были переведены в числовой формат {0, 1})
try:
    df['gender'] = df['gender'].apply(lambda x: x.replace('Male', '1').replace('Female', '0')).astype('int')
    df = df.rename(columns={'gender': 'male'})
except:
    pass
for item in boolean_features:
    try: 
        df[item] = df[item].apply(lambda x: x.replace('Yes', '1').replace('No', '0')).astype('int')
    except: 
        continue

df.reset_index(drop=True, inplace=True)
display(df)
print('**** numeric features:', *numeric_features, sep='\n')
print('_'*50)
print('**** categorical features:', *categorical_features, sep='\n')
```

### train_test_split (test_size = 0.25)

```python
X = df.drop('target', axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    stratify=y, 
                                                    test_size=TEST_SIZE,
                                                    random_state=RND)
```

### Scaler
Так как данные не распредлены нормально, выбор StandardScaler - не лучшая идея. Так как мы избавились от выбросов, можем использовать MaxAbsScaler

```python
scaler = Pipeline(steps=[('scaler', MaxAbsScaler())])
```

### Encoder

```python
onehot = Pipeline(steps=[('onehot', OHE(drop='first', handle_unknown='ignore'))])

ordinal = Pipeline(steps=[('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan))])

numeric_features = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X_train.select_dtypes(include=['object']).columns

# для линейных алгоритмов OHE
lr_preprocessor = ColumnTransformer(transformers=[('num', scaler, numeric_features),
                                                  ('cat', onehot, categorical_features)])
# для ансамблевых алгоритмов LabelEncoder
t_preprocessor = ColumnTransformer(transformers=[('num', scaler, numeric_features),
                                                 ('cat', ordinal, categorical_features)])
```

### Параметры для GridSearchSV

```python
skf = StratifiedKFold(n_splits=cv_split, 
                      shuffle=True,
                      random_state=RND)  

models = dict()

models['LR'] = {
    'preprocessor': [lr_preprocessor],
    'classifier': [LogisticRegression(random_state=RND)],
    'classifier__C': np.linspace(0.001, 99.001, 100),
    'classifier__solver': ['liblinear'],
    'classifier__penalty': ['l1', 'l2']
}

models['RF'] = {
    'preprocessor': [t_preprocessor],
    'classifier': [RandomForestClassifier(random_state=RND, n_jobs=-1)],
    'classifier__n_estimators': [20, 50, 100],
    'classifier__criterion': ['gini', 'entropy'],
    'classifier__max_depth': [5, 10, 15, 20],
    'classifier__min_samples_leaf': [1, 2, 4, 8, 10]
}

models['XGBoost'] = {
    'preprocessor': [t_preprocessor],
    'classifier': [XGBClassifier(n_jobs=-1, random_state=RND, n_estimators=500)],
    'classifier__gamma': [0, 0.1, 0.2, 0.3, 0.6],
    'classifier__max_depth': [2, 3, 4, 5, 6, 7, 8],
    'classifier__min_child_weight': [0.05, 0.1, 0.15, 0.2, 0.25, 0.3],
    'classifier__lambda': [0, 0.1, 0.2, 0.5, 0.6]
}
```

### Пайплайны моделей

```python
pipe = dict()

pipe['LR'] = Pipeline(steps=[("preprocessor", models['XGBoost']['preprocessor'][0]), 
                             ("classifier", models['XGBoost']['classifier'][0])])

pipe['RF'] =  Pipeline(steps=[("preprocessor", models['RF']['preprocessor'][0]), 
                              ("classifier", models['RF']['classifier'][0])])

pipe['XGBoost'] = Pipeline(steps=[("preprocessor", models['XGBoost']['preprocessor'][0]), 
                                  ("classifier", models['XGBoost']['classifier'][0])])
```

### Logistic Regression

```python
LR = GridSearchCV(pipe['LR'], models['LR'], scoring='roc_auc', cv=skf, n_jobs=-1)
LR.fit(X_train, y_train)
print('Logistic Regression (roc_auc_score):', LR.best_score_)
```

### Random Forest Classifier

```python
RF = GridSearchCV(pipe['RF'], models['RF'], scoring='roc_auc', cv=skf, n_jobs=-1)
RF.fit(X_train, y_train)
print('Random Forest (roc_auc_score):', RF.best_score_)
```

### XGBoost Classifier

```python
XGB = GridSearchCV(pipe['XGBoost'], models['XGBoost'], scoring='roc_auc', cv=skf, n_jobs=-1)
XGB.fit(X_train, y_train)
print('XGBoost Classifier (roc_auc_score):', XGB.best_score_)
```

### Анализ лучшей модели

Сравнивались 3 модели машинного обучения: линеная модель (логистическая регрессия), случайный лес и модель градиентного бустинга (XGBoost).

Лучшей моделью по метрике roc_auc_score оказалась модель градиентного бустинга XGBoost Classifier.
